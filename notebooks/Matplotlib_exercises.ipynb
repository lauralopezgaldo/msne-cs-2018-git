## Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
  
x = np.random.rand(150)
y = (-0.2)*np.random.rand(150) + 0.1
plt.scatter(x, y, color='r', s=np.random.randint(0, 10, size=(x.shape)), alpha=0.3)

plt.hold
x = 0.5 + np.random.rand(200)
y = -0.45*np.random.rand(200) + 0.20
plt.scatter(x, y, color='y', s=np.random.randint(0, 50, size=(x.shape)), alpha=0.2)

plt.hold
x = 1 + np.random.rand(200)
y = -0.8*np.random.rand(200) + 0.3
plt.scatter(x, y, color=(.5, 1, 0), s=np.random.randint(0, 75, size=(x.shape)), alpha=0.2)

plt.hold
x = 1.5 + np.random.rand(250)
y = -0.9*np.random.rand(250) +0.45
plt.scatter(x, y, color='c', s=np.random.randint(0, 100, size=(x.shape)), alpha=0.2)

plt.hold
x = 2 + np.random.rand(300)
y = -1.5*np.random.rand(300) + 0.75
plt.scatter(x, y, color='b', s=np.random.randint(0, 150, size=(x.shape)), alpha=0.2)

plt.hold
x = 2.5 + np.random.rand(300)
y = -2*np.random.rand(300) +1
plt.scatter(x, y, color='m', s=np.random.randint(0, 200, size=(x.shape)), alpha=0.2)

plt.xlabel('Colors', fontsize = 30)
plt.ylabel('Limits', fontsize = 20)

plt.xticks([0, 1.5, 2.5], ['RED', 'BLUE', 'PINK'], fontsize=20)
plt.yticks([1,0.66,0.33,0,-.33,-.66,-1],['6','4','2','0','-2','-4','-6'], fontsize=20)

plt.grid()
```
